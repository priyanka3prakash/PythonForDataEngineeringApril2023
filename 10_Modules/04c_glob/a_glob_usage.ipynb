```python
print("\n Current Directory")
all_files = glob.glob("*")
print(all_files)
```

```python
print("\n Current Directory - Only text files")
text_files = glob.glob("*.txt")
for each_txt_file in text_files:
    print("\t", each_txt_file)
```

```python
print("\n One Level Deep")
text_files = glob.glob("*/*.txt")
for each_txt_file in text_files:
    print("\t", each_txt_file)
```

```python
print("\n Two Levels Deep")
text_files = glob.glob("*/*/*.txt")
for each_txt_file in text_files:
    print("\t", each_txt_file)
```

```python
print("\n Three Levels Deep")
text_files = glob.glob("*/*/*/*.txt")
for each_txt_file in text_files[:10]:
    print("\t", each_txt_file)
```

```python
print("\n All levels Deep (as available)")
text_files = glob.glob("**/*.txt", recursive=True)
for each_txt_file in text_files[:10]:
    print("\t", each_txt_file)
```

```python
print("\n All levels Deep (as available) OF LICENSE files")
text_files = glob.glob("**/LICENSE.txt", recursive=True)
for each_txt_file in text_files[:10]:
    print("\t", each_txt_file)
```

```python
print("\n All levels Deep (as available) OF Particular Pattern")
text_files = glob.glob(r"Lib\site-packages\*-**/*", recursive=True)
for each_txt_file in text_files[:10]:
    print("\t", each_txt_file)
```

```python
# Getting a list of all files with multiple possible extensions
ext_files = glob.glob("*.[txt,csv,json]")
print(ext_files)
```

```python
# Getting a list of all files with multiple possible extensions
ext_files = glob.glob("**/*.[txt,csv,json]", recursive=True)
print(ext_files)
```

```python
# Getting a list of all files in a specific directory
dir_files = glob.glob("Lib/site-packages/*")
print(dir_files[:10])
```

```python
# Getting a list of all files in a directory and its subdirectories:
all_files = glob.glob("Lib/site-packages/**/*", recursive=True)
print(all_files)
```

```python
# Getting a list of all files in a directory sorted by modification time
files = glob.glob("lib/site-packages/*")
files = sorted(files, key=os.path.getmtime)

files
```

```python
# Getting a list of all files in a directory sorted by file size

files = glob.glob("lib/site-packages/*")
files = sorted(files, key=os.path.getsize)
```

```python
# Getting a list of all files in a directory sorted by file name

files = glob.glob("Lib/site-packages/*")
files = sorted(files)
```

```python
# Getting a list of all files in a directory sorted by file extension:
files = glob.glob("Lib/site-packages/*")
files = sorted(files, key=lambda x: os.path.splitext(x)[1])
```

```python
# Getting a list of all files in a directory that match a specific regular expression
import re

files = glob.glob("*")
regex = re.compile(r"^[a-zA-Z]{3}_\d{4}\.txt$")
matched_files = list(filter(regex.match, files))
print(matched_files)
```
