```python
def scan_files_by_extn(given_file_etxn, _path):
    files_with_given_extn = []
    for ech_dir, list_of_dirs, list_of_file in os.walk(_path):
        # print(ech_dir, list_of_dirs, list_of_file)
        #print(list_of_file)
        for ech_file in list_of_file:
            file_etxn = os.path.splitext(ech_file)[1]
            if file_etxn == given_file_etxn:
                files_with_given_extn.append(ech_file)
    
    return files_with_given_extn

scan_files_by_extn(".py", "/workspaces/PythonForDataEngineeringApril2023/" )
```

```python
def scan_files_by_extn(given_file_etxn, _path):
    files_with_given_extn = []
    for ech_dir, list_of_dirs, list_of_file in os.walk(_path):
        # print(ech_dir, list_of_dirs, list_of_file)
        #print(list_of_file)
        for ech_file in list_of_file:
            file_etxn = os.path.splitext(ech_file)[1]
            if file_etxn == given_file_etxn:
                files_with_given_extn.append(
                    os.path.join(ech_dir, ech_file )
                )
    
    return files_with_given_extn

scan_files_by_extn(".md", "/workspaces/PythonForDataEngineeringApril2023/" )
```

## Filter the files based on size

```python
import sys 

sys.exec_prefix
```

```python
def get_files_greater_than(_path, size=1024): # 1kb = 1024
    for each_dir, list_of_dirs, list_of_files in os.walk(_path):
        for filename in list_of_files:
            print(filename, os.path.getsize(filename))


get_files_greater_than(".")
```

```python
def get_files_greater_than(_path, size=1024): # 1kb = 1024
    for each_dir, list_of_dirs, list_of_files in os.walk(_path):
        for filename in list_of_files:
            file_path = os.path.join(each_dir, filename)
            if os.path.getsize(file_path) > size:
                print(filename, os.path.getsize(file_path))


get_files_greater_than(sys.exec_prefix)
```

```python
def get_files_greater_than(_path, size=1024): # 1kb = 1024
    for eachobj in os.scandir(_path):
        if eachobj.is_file:
            print(eachobj.stat().st_size, eachobj.name)


get_files_greater_than(sys.exec_prefix)
```
